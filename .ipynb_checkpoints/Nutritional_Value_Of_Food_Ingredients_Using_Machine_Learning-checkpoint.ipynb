```python
data = pd.get_dummies(data, columns=['ingredients', 'diet', 'flavor_profile'], dtype=int)
print(data)
```

```python
target= data['Target']
data= data.drop(columns=['Target','course','state','region','food_name','percentage_of_ingredients','units'],axis=1)
data
```

```python
print("\nDataFrame (df) without 'course', 'state', 'region', 'food_name':\n", data)
print("\nTarget (Target):\n", target)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(target)
print("\nLabel Encoded Target:\n", label)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(data,label, test_size=0.3)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(data, label, test_size=0.3, random_state=42)
```

```python
# Train a model (example: Logistic Regression) with increased max_iter
model = LogisticRegression(max_iter=1000)  # Increasing max_iter to 1000
model.fit(X_train, Y_train)
```

```python
ypred = model.predict(X_test)
```

```python
# Generate the classification report
report = classification_report(Y_test, ypred, zero_division=1, output_dict=True)
print(classification_report(Y_test, ypred, zero_division=1))
```

```python
# Calculate the accuracy
accuracy = model.score(X_test, Y_test)
print(f"Accuracy: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
labels = list(report.keys())[:-3] 
precision = [report[label]['precision'] for label in labels]
recall = [report[label]['recall'] for label in labels]
f1 = [report[label]['f1-score'] for label in labels]
labels.append('accuracy')
precision.append(0)  
recall.append(0)    
f1.append(0)        
x = np.arange(len(labels))  
width = 0.2 
fig, ax = plt.subplots(figsize=(14, 6))
rects1 = ax.bar(x - width, precision, width, label='Precision')
rects2 = ax.bar(x, recall, width, label='Recall')
rects3 = ax.bar(x + width, f1, width, label='F1-score')
rects4 = ax.bar(len(labels) - 1, accuracy, width, label='Accuracy', color='cyan')
ax.set_xlabel('Labels')
ax.set_ylabel('Scores')
ax.set_title('Precision, Recall, F1-score, and Accuracy by class')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4) 

fig.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(data, label, test_size=0.4, random_state=42)
```

```python
for k in range(1, 20):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, Y_train)
    accuracy = model.score(X_test, Y_test)
    print(f'K: {k}\tAccuracy:{accuracy:.2%}')
```

```python
model = KNeighborsClassifier(n_neighbors=7)  # You can adjust n_neighbors if needed
model.fit(X_train, Y_train)
```
